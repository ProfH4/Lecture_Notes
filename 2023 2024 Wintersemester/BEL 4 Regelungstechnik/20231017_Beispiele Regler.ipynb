## P-Regler Beispiel

Ein proportionaler Regelkreis (P-Regler) ist ein grundlegendes Konzept der Regelungstechnik. Er wird oft verwendet, um ein System auf einen gew√ºnschten Sollwert zu regeln, indem er proportional zum Fehler zwischen dem Ist- und dem Sollwert arbeitet.

Hier ist ein einfacher Python-Code, der ein System mit einem P-Regler simuliert:

```python
import matplotlib.pyplot as plt

def p_regler(kp, setpoint, initial_value, time_steps):
    time = [0]
    value = [initial_value]
    setpoint_values = [setpoint]
    error_values = [setpoint - initial_value]
    
    for i in range(1, time_steps+1):
        error = setpoint - value[-1]
        control_signal = kp * error
        value.append(value[-1] + control_signal)
        time.append(i)
        setpoint_values.append(setpoint)
        error_values.append(error)
    
    return time, value, setpoint_values, error_values

# Parameter
kp = 0.1  # Proportionalit√§tskonstante
setpoint = 10  # Sollwert
initial_value = 0  # Anfangswert
time_steps = 200  # Anzahl der Zeitschritte

# Simuliere den P-Regler
time, output, setpoint_values, error_values = p_regler(kp, setpoint, initial_value, time_steps)

# Plot Eingang (Sollwert und Fehler)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(time, setpoint_values, label='Sollwert', color='red')
plt.plot(time, error_values, label='Fehler', color='blue')
plt.xlabel('Zeitschritte')
plt.ylabel('Wert')
plt.title('Eingang des P-Reglers')
plt.legend()

# Plot Ausgang
plt.subplot(1, 2, 2)
plt.plot(time, output, label='Ausgang', color='green')
plt.xlabel('Zeitschritte')
plt.ylabel('Wert')
plt.title('Ausgang des P-Reglers')
plt.legend()

plt.tight_layout()
plt.show()
```

# D-Regler Beispiel

Ein Differentialregler (D-Regler) reagiert proportional zur zeitlichen √Ñnderung des Fehlers. Es ist wichtig zu beachten, dass ein reiner D-Regler oft instabil ist, daher wird er in der Praxis oft mit einem P- oder PI-Regler kombiniert, um ein PD- oder PID-Regler zu erhalten.

Hier ist ein einfacher Python-Code, der ein System mit einem D-Regler simuliert:

```python
import matplotlib.pyplot as plt

def d_regler(control_signal_func, setpoint_func, initial_value, time_steps):
    time = [0]
    value = [initial_value]
    setpoint_values = [setpoint_func(0)]
    error_values = [setpoint_func(0) - initial_value]
    
    for i in range(1, time_steps+1):
        error = setpoint_func(i-1) - value[-1]
        control_signal = control_signal_func(i-1, error)
        value.append(value[-1] + control_signal)
        time.append(i)
        setpoint_values.append(setpoint_func(i))
        error_values.append(error)
    
    return time, value, setpoint_values, error_values

# Define time-varying control signal and setpoint functions
def control_signal_func(t, error):
    return 0.1 * t * error  # Example: Control signal varies with time

def setpoint_func(t):
    return 10 + 0.2 * t  # Example: Setpoint increases linearly with time

# Parameter
initial_value = 0  # Anfangswert
time_steps = 50  # Anzahl der Zeitschritte

# Simuliere den D-Regler
time, output, setpoint_values, error_values = d_regler(control_signal_func, setpoint_func, initial_value, time_steps)

# Plot Eingang (Sollwert und Fehler)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(time, setpoint_values, label='Sollwert', color='red')
plt.plot(time, error_values, label='Fehler', color='blue')
plt.xlabel('Zeitschritte')
plt.ylabel('Wert')
plt.title('Eingang des D-Reglers')
plt.legend()

# Plot Ausgang
plt.subplot(1, 2, 2)
plt.plot(time, output, label='Ausgang', color='green')
plt.xlabel('Zeitschritte')
plt.ylabel('Wert')
plt.title('Ausgang des D-Reglers')
plt.legend()

plt.tight_layout()
plt.show()
```

# PID Regler Beispiel

Ein PID-Regler kombiniert proportional, integral und differential Aktionen, um eine genauere Regelung zu erm√∂glichen.
