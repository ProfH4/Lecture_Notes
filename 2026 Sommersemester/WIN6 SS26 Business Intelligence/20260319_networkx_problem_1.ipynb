```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6])

G.add_edges_from([(1,2), (1,3), (2,4), (3,4), (4,5), (5,6)])

nx.draw(G)
```

```python
A = nx.adjacency_matrix(G)
print(A.toarray())
```
